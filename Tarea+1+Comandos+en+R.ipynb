**PREGUNTA 6:los siguientes son una muestra de observaciones sobre la radiacion solar entrante en un invernadero:
11.1 10.6 6.3 8.8 10.7 11.2 8.9 12.2

Asignamos los datos a un objeto solar radiación.

```r
solar.radiacion = c(11.1, 10.6, 6.3, 8.8, 10.7, 11.2, 8.9, 12.2)
solar.radiacion
```

```r
Media,mediana y varianza:
```

```r
mean(solar.radiacion) #nos muestra la media de los datos
```

```r
median(solar.radiacion) #nos muestra la mediana de los datos
```

```r
sd(solar.radiacion) #nos muestra la varianza de los datos
```

Agregar 10 a cada dato de solar.radiacion y asigna el resultado a sr10. Encontrar la media,mediana ya varianza de sr10. ¿Como la estadistica cambia?

```r
sr = solar.radiacion + c(10,10)
sr
```

```r
mean(sr)
```

```r
#notamos que el valor de la media varia exactamente en 10, que es lo que aumentamos a cada valor
```

```r
median(sr)
```

```r
#igualmente la mediana aumentó en 10 
```

```r
sd(sr)
```

```r
#la varianza mantiene su valor 
```

Ahora multiplicaremos cada observacion por -2 y le asignaremos el valor srm2. Encontrar la media,mediana ya varianza de srm2. ¿Como la estadistica cambia?

```r
srm2 = solar.radiacion * -2
srm2
```

```r
mean(srm2)
```

```r
median(srm2)
sd(srm2)
```
